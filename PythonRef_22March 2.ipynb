```python
a = [1,2,4,3]
b=[]

for i in a:
  if i%2==0:
    b.append(i)
  else:
    b.append(i**2)

print(b)
```

```python
[i if i%2==0 else i**2 for i in a]
```

```python
#set comprehension
{i**2 for i in range(1,11)}
```

```python
#tuple comprehension
tuple(i**2 for i in range(1,11))
```

```python
#dictionary comprehension
{i:i**2 for i in range(1,11)}
```
